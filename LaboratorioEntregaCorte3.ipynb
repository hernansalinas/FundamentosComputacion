# Comandos de interés de pandas

| Comando                      | Explicación                                                                                          |
|------------------------------|-----------------------------------------------------------------------------------------------------|
| `pd.DataFrame(data)`         | Crea un DataFrame a partir de un diccionario de datos o de una lista de listas, diccionarios, etc.  |
| `df['columna']`              | Selecciona una columna específica de un DataFrame.                                                  |
| `df[['col1', 'col2']]`       | Selecciona múltiples columnas específicas de un DataFrame.                                          |
| `df.loc[índice]`             | Selecciona una fila específica por índice, utilizando la etiqueta del índice.                      |
| `df.iloc[índice]`            | Selecciona una fila específica por índice, utilizando la posición numérica del índice.              |
| `df[df['columna'] > valor]`  | Filtra las filas del DataFrame donde los valores de una columna son mayores que un valor dado.      |
| `df['columna'].mean()`       | Calcula la media (promedio) de una columna numérica.                                               |
| `df['columna'].std()`        | Calcula la desviación estándar de una columna numérica.                                            |
| `df['columna'].min()`        | Encuentra el valor mínimo en una columna.                                                          |
| `df['columna'].max()`        | Encuentra el valor máximo en una columna.                                                          |
| `df.describe()`              | Genera estadísticas descriptivas para columnas numéricas, como la media, la desviación estándar, etc. |
| `df[df['columna'].notnull()]`| Filtra las filas del DataFrame donde los valores de una columna no son nulos.                      |
| `df.groupby('columna')`      | Agrupa el DataFrame por una o más columnas.                                                        |
| `df.groupby('columna').mean()` | Calcula la media de cada grupo para columnas numéricas después de agrupar por una columna.        |
| `df.plot(kind='bar')`        | Crea un gráfico de barras para los datos de un DataFrame.                                          |
| `df.plot(kind='scatter', x='col1', y='col2')` | Crea un gráfico de dispersión utilizando dos columnas especificadas.                   |
| `df.apply(func, axis=1)`     | Aplica una función a lo largo de un eje del DataFrame (por filas o columnas).                      |
| `df['columna'].astype(tipo)` | Cambia el tipo de datos de una columna específica.                                                 |
| `plt.xlabel('texto')`        | Establece la etiqueta del eje x en una gráfica de Matplotlib.                                      |
| `plt.ylabel('texto')`        | Establece la etiqueta del eje y en una gráfica de Matplotlib.                                      |
| `plt.title('texto')`         | Establece el título de una gráfica de Matplotlib.                                                  |
| `plt.grid(True)`             | Activa el grid (rejilla) en una gráfica de Matplotlib.                                             |
| `plt.show()`                 | Muestra la gráfica generada en Matplotlib.                                                         |

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from io import StringIO
import seaborn as sns
sns.set_theme(style="darkgrid")
```

```python
# Base de datos
data = """
 Hidrógeno | H | 1 | 1.008 | 14.01 | 20.28 | 2.20
 Helio | He | 2 | 4.0026 | 0.95 | 4.22 | -
 Litio | Li | 3 | 6.94 | 453.69 | 1560 | 0.98
 Berilio | Be | 4 | 9.0122 | 1560 | 2742 | 1.57
 Boro | B | 5 | 10.81 | 2349 | 4200 | 2.04
 Carbono | C | 6 | 12.011 | 3823 | 4098 | 2.55
 Nitrógeno | N | 7 | 14.007 | 63.15 | 77.36 | 3.04
 Oxígeno | O | 8 | 15.999 | 54.36 | 90.20 | 3.44
 Flúor | F | 9 | 18.998 | 53.53 | 85.03 | 3.98
 Neón | Ne | 10 | 20.180 | 24.56 | 27.07 | - """

# Leer base de datos
elementos_df = pd.read_csv(StringIO(data), sep="|", header=None)
elementos_df.columns = ["Elemento", "Símbolo", "Número Atómico", "Masa Atómica (u)", "Punto de Fusión (K)", "Punto de Ebullición (K)", "Electronegatividad (Pauling)"]
elementos_df.columns = elementos_df.columns.str.replace(' ','_').str.replace('(','').str.replace(')','').str.lower().str.replace('á','a').str.replace('é','e').str.replace('í','i').str.replace('ó','o').str.replace('ú','u')
elementos_df
#Utiliza los datos proporcionados para crear un DataFrame llamado df_elementos.
```

```python
T_Ambiente_Med = 298.15 #Temperatura ambiente promedio Kelvin

def estado_de_la_materia(punto_de_fusion_k, punto_de_ebullicion_k):

  if T_Ambiente_Med < punto_de_fusion_k:
    return "solido"
  elif T_Ambiente_Med > punto_de_ebullicion_k:
    return "gas"
  else:
    return "líquido"

elementos_df['estado_de_la_materia'] = elementos_df.apply(lambda row: estado_de_la_materia(row['punto_de_fusion_k'], row['punto_de_ebullicion_k']), axis=1)

elementos_gas = elementos_df[elementos_df['estado_de_la_materia'] == "gas"]
elementos_gas

#Agrupa los elementos por estado de la materia a T_Ambiente_Med ambiente (solido, líquido, gas) basado en los puntos de fusión y ebullición.
```

```python
elementos_solido = elementos_df[elementos_df['estado_de_la_materia'] == "solido"]
elementos_solido
```

```python
elementos_liquido = elementos_df[elementos_df['estado_de_la_materia'] == "liquido"]
elementos_liquido
```

```python
mean_solido = elementos_solido['masa_atomica_u'].mean()

mean_gas = elementos_gas['masa_atomica_u'].mean()

mean_liquido = elementos_liquido['masa_atomica_u'].mean()

print("Masa atómica promedio de los elementos:")
print("Solido:", mean_solido)
print("Gas:", mean_gas)
print("Líquido:", mean_liquido)
#Calcula la masa atómica promedio para cada grupo de estado de la materia
```

```python
promedios = mean_solido, mean_gas, mean_liquido

estados = ['Sólido', 'Gas', 'Líquido']
promedios = [mean_solido, mean_gas, mean_liquido]
plt.bar(estados, promedios)

plt.xlabel('Estado de la Materia')
plt.ylabel('Masa Atómica Promedio (u)')
plt.title('Comparación de Masas Atómicas Promedio por Estado de la Materia')
plt.show()

#Realiza una visualización que compare las masas atómicas promedio de los diferentes estados
```

```python
# Crear el gráfico de dispersión
plt.figure(figsize=(10, 6))
scatter = plt.scatter(df_particulas['masa_mev/c^2'], df_particulas['tiempo_de_vida_media_s'],
                      c=range(len(df_particulas)), cmap='viridis')

# Añadir etiquetas a cada punto
for i, row in df_particulas.iterrows():
    plt.annotate(row['particula'], (row['masa_mev/c^2'], row['tiempo_de_vida_media_s']), fontsize=9, ha='right')

# Configurar ejes y título
plt.xlabel('Masa (MeV/c^2)')
plt.ylabel('Tiempo de Vida Media (s)')
plt.title('Masa vs Tiempo de Vida Media de las Partículas')
plt.yscale('log')
plt.xscale('log')
plt.xlim(1e-3, 1e4)  # Establecer límites del eje x
plt.ylim(1e-20, 1e10)  # Establecer límites del eje y
plt.grid()
plt.tight_layout()
plt.show()
```
