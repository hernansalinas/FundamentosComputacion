```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Punto 1
ctes_agua_salada={
    "rho":1205,
    "g":9.81,
    "P0":101325,
}

profundidades=np.arange(0,42,2)
presiones= ctes_agua_salada["P0"]+(ctes_agua_salada["rho"]*ctes_agua_salada["g"]*profundidades)

#Punto 2
df_presion=pd.DataFrame({"Profundidad[m]":profundidades,"PresionSalada[Pa]":presiones})
print(df_presion)

#punto 3
ctes_agua_ducle={
    "rho":1000,
    "g":9.81,
    "P0":101325,
}

#Punto 4
presionesDulce= ctes_agua_ducle["P0"]+(ctes_agua_ducle["rho"]*ctes_agua_ducle["g"]*profundidades)

df_presion["PresionDulce[Pa]"]=presionesDulce
print(df_presion)

#Punto 5
plt.plot(df_presion["Profundidad[m]"],df_presion["PresionSalada[Pa]"])
plt.plot(df_presion["Profundidad[m]"],df_presion["PresionDulce[Pa]"])
plt.xlabel('Profundidad[m]')
plt.ylabel('Presion[Pa]')
plt.title('Presion vs Profundidad')
plt.legend(['PresionSalada[Pa]','PresionDulce[Pa]'])
plt.show()

#Punto 7
df_filtro_presion = df_presion[(df_presion["Profundidad[m]"] > 20) & (df_presion["Profundidad[m]"] < 30)]
print(df_filtro_presion)

#Punto 8
df_presion.to_csv("presion.csv")
```

La presion con respecto a la profundidad, aumenta progresivamente y de forma lineal.
Sobre la teoria, el agua salada al ser mas densa, incrementa mas velozmente la presion, efecto que vemos expresado en la grafica, en la que el incremento de la presion en el agua salada es mayor que el del agua dulce con respecto a la profundidad
