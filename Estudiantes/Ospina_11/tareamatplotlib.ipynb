```python
calificaciones = {
    'Ana': 85,
    'Juan': 92,
    'María': 78,
    'Pedro': 88,
    'Lucía': 95
}
calificaciones['Carlos']=80
calificaciones.pop(min(calificaciones, key=calificaciones.get))
calificaciones['María'] = 82
```

```python
def productos_extremos(productos):
    producto_mas_caro = max(productos, key=productos.get)
    producto_mas_barato = min(productos, key=productos.get)
    return producto_mas_caro, producto_mas_barato

productos = {
    'manzanas': 2.5,
    'naranjas': 3.0,
    'plátanos': 1.2,
    'peras': 2.8
}

producto_caro, producto_barato = productos_extremos(productos)
print(f"El producto más caro es: {producto_caro}")
print(f"El producto más barato es: {producto_barato}")
```

```python
import pandas as pd
serie_calificaciones = pd.Series(calificaciones)

print(serie_calificaciones)

promedio = serie_calificaciones.mean()

calificaciones_mayores_85 = serie_calificaciones[serie_calificaciones > 85]

print(f"Calificación promedio: {promedio}")
print("Calificaciones mayores a 85:")
print(calificaciones_mayores_85)
```

```python
import pandas as pd

data = {
    'Nombre': ['Ana', 'Juan', 'María', 'Pedro', 'Lucía'],
    'Edad': [16, 17, 15, 16, 17],
    'Grado': [10, 11, 10, 11, 12]
}

df_estudiantes = pd.DataFrame(data)

print(df_estudiantes)


filtro = (df_estudiantes['Grado'] == 11) & (df_estudiantes['Edad'] >= 16)
estudiantes_filtrados = df_estudiantes[filtro]

resultado = estudiantes_filtrados[['Nombre', 'Edad']]

print(resultado)
```

```python
import pandas as pd

data = {
    'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
            'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
    'Ventas': [2000, 2100, 1800, 2400, 2300, 2200, 2500, 2700,
               2600, 2900, 3000, 3100]
}

df_ventas = pd.DataFrame(data)

mes_max_ventas = df_ventas.loc[df_ventas['Ventas'].idxmax()]
mes_min_ventas = df_ventas.loc[df_ventas['Ventas'].idxmin()]
venta_promedio = df_ventas['Ventas'].mean()

print(f"Mes con la mayor cantidad de ventas: {mes_max_ventas['Mes']} ({mes_max_ventas['Ventas']} unidades)")
print(f"Mes con la menor cantidad de ventas: {mes_min_ventas['Mes']} ({mes_min_ventas['Ventas']} unidades)")
print(f"Venta promedio mensual: {venta_promedio:.2f} unidades")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

planetas = {
    'Planeta': ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno'],
    'Masa (10^24 kg)': [0.330, 4.87, 5.97, 0.642, 1898, 568, 86.8, 102],
    'Radio (km)': [2439.7, 6051.8, 6371, 3389.5, 69911, 58232, 25362, 24622],
    'Distancia al Sol (10^6 km)': [57.9, 108.2, 149.6, 227.9, 778.5, 1434, 2871, 4495],
    'Periodo Orbital (días)': [88, 224.7, 365.2, 687, 4331, 10747, 30589, 59800],
    'Gravedad (m/s^2)': [3.7, 8.9, 9.8, 3.7, 24.8, 10.4, 8.7, 11.2]
}

df_planetas = pd.DataFrame(planetas)

planetas_gravedad_alta = df_planetas[df_planetas['Gravedad (m/s^2)'] > 10]
planeta_periodo_max = df_planetas.loc[df_planetas['Periodo Orbital (días)'].idxmax()]
masa_promedio = df_planetas['Masa (10^24 kg)'].mean()

print("Planetas con gravedad mayor a 10 m/s²:")
print(planetas_gravedad_alta[['Planeta', 'Gravedad (m/s^2)']])
print(f"\nPlaneta con el periodo orbital más largo: {planeta_periodo_max['Planeta']} ({planeta_periodo_max['Periodo Orbital (días)']} días)")
print(f"\nPromedio de la masa de los planetas: {masa_promedio:.2f} x 10^24 kg")


```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_planetas['Radio (km)'], df_planetas['Masa (10^24 kg)'], color='blue', s=100, edgecolor='k')

plt.xlabel('Radio (km)')
plt.ylabel('Masa (10^24 kg)')
plt.title('Masa de los Planetas en Función de su Radio')
for i, planeta in enumerate(df_planetas['Planeta']):
    plt.text(df_planetas['Radio (km)'][i], df_planetas['Masa (10^24 kg)'][i], planeta, fontsize=9, ha='right')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.scatter(df_planetas['Periodo Orbital (días)'], df_planetas['Distancia al Sol (10^6 km)'], color='green', s=100, edgecolor='k')

plt.xlabel('Periodo Orbital (días)')
plt.ylabel('Distancia al Sol (10^6 km)')
plt.title('Distancia al Sol en Función del Periodo Orbital')

for i, planeta in enumerate(df_planetas['Planeta']):
    plt.text(df_planetas['Periodo Orbital (días)'][i], df_planetas['Distancia al Sol (10^6 km)'][i], planeta, fontsize=9, ha='right')

plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
df_planetas_sorted = df_planetas.sort_values(by='Gravedad (m/s^2)', ascending=False)
plt.bar(df_planetas_sorted['Planeta'], df_planetas_sorted['Gravedad (m/s^2)'], color='orange', edgecolor='k')

plt.xlabel('Planeta')
plt.ylabel('Gravedad (m/s^2)')
plt.title('Gravedad Superficial de los Planetas')

plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
correlation_matrix = df_planetas[['Masa (10^24 kg)', 'Radio (km)', 'Distancia al Sol (10^6 km)', 'Periodo Orbital (días)', 'Gravedad (m/s^2)']].corr()

print("Matriz de Correlación:")
print(correlation_matrix)
```
