#Parte 1 Solución


````python
#CALIFICACIONESS

calificaciones = {
    'David': 85,
    'Pepito': 92,
    'María': 78,
    'Perez': 88,
    'Lucía': 95
}

#Add
calificaciones["Carlos"]=80
print("With Carlos:", calificaciones)
#Minimum
min_cal = min(calificaciones, key=calificaciones.get)
print("Minimum:", min_cal)
#Delete
del calificaciones[min_cal]
print("Without minimum:", calificaciones)
#Update
calificaciones["María"]=82

print("Maria update:", calificaciones)
````

````python
#PRODUCTOS

productos = {
    'manzanas': 2.5,
    'naranjas': 3.0,
    'plátanos': 1.2,
    'peras': 2.8
}

#These functions (max and min) already exist

def caro(productos):
    producto_caro = None
    precio_mas_alto = float('-inf')  # Valor más bajo posible

    for producto, precio in productos.items():
        if precio > precio_mas_alto:
            precio_mas_alto = precio
            producto_caro = producto

    return producto_caro

def barato(productos):
    producto_barato = None
    precio_mas_bajo = float('inf')   # Valor más alto posible

    for producto, precio in productos.items():
        if precio < precio_mas_bajo:
            precio_mas_bajo = precio
            producto_barato = producto

    return producto_barato

print("El producto más caro es:", caro(productos))
print("El producto más barato es:", barato(productos))
````

````python
import pandas as pd
import matplotlib.pyplot as plt


planetas = {
    "Planeta": ["Mercurio", "Venus", "Tierra", "Marte", "Júpiter", "Saturno", "Urano", "Neptuno"],
    "Masa (10^24 kg)": [0.330, 4.87, 5.97, 0.642, 1898, 568, 86.8, 102],
    "Radio (km)": [2439, 6052, 6371, 3390, 71492, 60268, 25559, 24764],
    "Distancia al Sol (10^6 km)": [57.9, 108.2, 149.6, 227.9, 778.5, 1426.7, 2872.5, 4495.1],
    "Periodo Orbital (dias)": [88, 224.7, 365.2, 687, 4333, 10759, 30687, 60190],
    "Gravedad (m/s^2)": [3.7, 8.87, 9.81, 3.71, 24.79, 10.44, 8.87, 11.15]
}

df_planetas = pd.DataFrame(planetas)
print("-------------------- df --------------------")
print(df_planetas)

#4.2 Utiliza Matplotlib para crear una gráfica de dispersión (scatter plot) que muestre la masa de cada planeta en función de su radio. Etiqueta los ejes de la gráfica adecuadamente.

print("+-" + "-" * 30 + "-+")
plt.scatter(df_planetas["Radio (km)"], df_planetas["Masa (10^24 kg)"])

#to get the names of the planets shown
for i, row in df_planetas.iterrows():
    plt.annotate(row['Planeta'], (row['Radio (km)'], row['Masa (10^24 kg)']))

plt.xlabel("Radio (km)")
plt.ylabel("Masa (10^24 kg)")
plt.title("Relación entre Radio y Masa de los Planetas")
plt.show()

#4.3 Crea una gráfica de dispersión que muestre la distancia de cada planeta al Sol en función de su periodo orbital. Identifica si existe una relación entre estas dos variables.
print("+-" + "-" * 30 + "-+")
plt.scatter(df_planetas["Periodo Orbital (dias)"], df_planetas["Distancia al Sol (10^6 km)"])

#to get the names of the planets shown
for i, row in df_planetas.iterrows():
    plt.annotate(row['Planeta'], (row["Periodo Orbital (dias)"], row["Distancia al Sol (10^6 km)"]))

plt.xlabel("Periodo Orbital (dias)")
plt.ylabel("Distancia al Sol (10^6 km)")
plt.title("Distancia de cada planeta al Sol en función de su periodo orbital")
plt.show()

print("Relación: mayor distancia al sol, mayor periodo orbital")


#4.4 Crea un gráfico de barras que muestre la gravedad superficial de cada planeta. Ordena los planetas en el gráfico de mayor a menor gravedad.
print("+-" + "-" * 30 + "-+")

#fix the df to have a descending order
df_planetas = df_planetas.sort_values(by="Gravedad (m/s^2)", ascending=False)

plt.bar(df_planetas["Planeta"], df_planetas["Gravedad (m/s^2)"])
plt.xlabel("Planeta")
plt.ylabel("Gravedad (m/s^2)")
plt.title("Gravedad superficial de cada planeta")
plt.show()


#4.5 Calcula la matriz de correlación para las columnas numéricas del DataFrame. Identifica las relaciones más fuertes entre las diferentes características de los planetas.
print("+-" + "-" * 30 + "-+")
#select only the numerical columns
import numpy as np
numeric_cols = df_planetas.select_dtypes(include=np.number)

#calculate the correlation matrix
correlation_matrix = numeric_cols.corr()

print("-------------------- Matriz de correlación --------------------")
print(correlation_matrix)

#find the pairs with the highest absolute correlation
top_correlations = correlation_matrix.unstack().sort_values(ascending=False)
print("-------------------- Correlaciones fuertes --------------------")
print(top_correlations.head(10))
````
