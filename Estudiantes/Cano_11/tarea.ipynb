```python
r={'manzanas': 2.5,
    'naranjas': 3.0,
    'plátanos': 1.2,
    'peras': 2.8}

def precios(lista):
  precio_Caro=0
  precio_barato=100000
  for x,y in lista.items():
    if y > precio_Caro:
      precio_Caro=y
    if y < precio_barato:
      precio_barato=y
  print('producto mas caro es:',precio_Caro)
  print('producto mas barato es:',precio_barato)
  return precio_Caro,precio_barato
precios(r)





```

```python
t={'Mes': [
        'Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio',
        'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'
    ],
    'Ventas': [
        2000, 2100, 1800, 2400, 2300, 2200,
        2500, 2700, 2600, 2900, 3000, 3100
    ]}
df_ventas=pd.DataFrame(t)
df_ventas
```

```python
print('Mes con ventas mas altas')
d=df_ventas['Ventas'].max()
df_ventas[df_ventas['Ventas']==d]
```

```python
print('Mes con ventas mas bajas')
dd=df_ventas['Ventas'].min()
df_ventas[df_ventas['Ventas']==dd]
```

```python
u={'Planeta': ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno'],
    'Masa (10^24 kg)': [0.330, 4.87, 5.97, 0.642, 1898, 568, 86.8, 102],
    'Radio (km)': [2439.7, 6051.8, 6371, 3389.5, 69911, 58232, 25362, 24622],
    'Distancia al Sol (10^6 km)': [57.9, 108.2, 149.6, 227.9, 778.5, 1434, 2871, 4495],
    'Periodo Orbital (días)': [88, 224.7, 365.2, 687, 4331, 10747, 30589, 59800],
    'Gravedad (m/s^2)': [3.7, 8.9, 9.8, 3.7, 24.8, 10.4, 8.7, 11.2]}
df_planetas=pd.DataFrame(u)
df_planetas
```

```python
plt.figure(figsize=(5,5))
df_planetas.plot(x='Masa (10^24 kg)',y='Radio (km)',kind='scatter')
plt.show()
```

```python
plt.figure(figsize=(5,5))
df_planetas.plot(x='Distancia al Sol (10^6 km)',y='Periodo Orbital (días)',kind='scatter')
plt.show()
```

```python
df_ordenado = df_planetas.sort_values(by='Gravedad (m/s^2)', ascending=False)

plt.figure(figsize=(5,5))
df_ordenado.set_index('Planeta',inplace=True)
df_ordenado.plot(y='Gravedad (m/s^2)',kind='bar')
plt.show()
```
