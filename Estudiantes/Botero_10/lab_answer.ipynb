```python
promd_masas= df_elementos["Masa Atomica(u)"].mean()
print(f"El promedio de las masas atomicas es: {promd_masas}")
```

```python
minima_fusion= df_elementos["Punto de Fusion(k)"].min()
m_fusion1= df_elementos[df_elementos["Punto de Fusion(k)"]==minima_fusion]["Elemento"].values[0]
print(f"El elemento que tiene un punto de fusión mas bajo es:  {m_fusion1}")


maxima_fusion= df_elementos["Punto de Fusion(k)"].max()
m_fusion1= df_elementos[df_elementos["Punto de Fusion(k)"]==maxima_fusion]["Elemento"].values[0]
print(f"El elemento que tiene un punto de fusión mas alto es:  {m_fusion1}")
```

```python
promedio_e= df_elementos["Electronegativida"].mean()
print(f"El promedio de las electronegatividades es: {promedio_e}")
desviacion_e = df_elementos["Electronegativida"].std()
print(f"La desviación estandar de las electronegatividades es: {desviacion_e}")
```

```python
masa_promd=df_elementos.groupby("Estado en temperatura ambiente (293 K)")["Masa Atomica(u)"].mean()
print(masa_promd)
```

```python
#2
df_particulas= df_particulas[df_particulas["Tiempo de Vida Media (s)"] != "Estable"]
df_particulas["Tiempo de Vida Media (s)"]= pd.to_numeric(df_particulas["Tiempo de Vida Media (s)"])
plt.figure(figsize=(5, 3))
plt.scatter(df_particulas["Tiempo de Vida Media (s)"], df_particulas["Masa (MeV/c^2)"])
plt.xscale("log")
plt.title("Masa versus tiempo")
plt.grid(axis="y", alpha=0.7)
plt.show()
```

```python
#3
c = {"Fuerte": "blue","Electromagnética": "red","Electromagnética, Fuerte": "purple","Fuerte, Débil": "yellow"}
df_particulas["Color"]= df_particulas["Interacción fundamental"].map(c)
plt.figure(figsize=(5, 3))
for i, color in c.items():
    subset= df_particulas[df_particulas['Interacción fundamental'].str.contains(i)]
    plt.scatter(subset['Masa (MeV/c^2)'], subset['Carga(e)'], label=i, color=color)

plt.title("Carga de Partículas segun la masa")
plt.legend(title="interacción Fundamental")
plt.grid(axis="y", alpha=0.7)
plt.show()
```

```python
#4
spin= df_particulas["Spin"].value_counts()

spin.plot(kind="bar") #La funcion "bar" es para que el grafico sea de barras
plt.title("Número de Partículas por Spin")
plt.grid(axis="y", alpha=0.7)
plt.show()
```
