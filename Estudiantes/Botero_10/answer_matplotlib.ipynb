```python
#3.3
v1= {"Mes":"Enero", "Ventas":2000}
v2= {"Mes":"Febrero", "Ventas":2100}
v3= {"Mes":"Marzo", "Ventas":1800}
v4= {"Mes":"Abril", "Ventas":2400}
v5= {"Mes":"Mayo", "Ventas":2300}
v6= {"Mes":"Junio", "Ventas":2200}
v7= {"Mes":"Julio", "Ventas":2500}
v8= {"Mes":"Agosto", "Ventas":2700}
v9= {"Mes":"Septiembre", "Ventas":2600}
v10= {"Mes":"Octubre", "Ventas":2900}
v11= {"Mes":"Noviembre", "Ventas":3000}
v12= {"Mes":"Diciembre", "Ventas":3100}
ms1=pd.Series(v1)
ms2=pd.Series(v2)
ms3=pd.Series(v3)
ms4=pd.Series(v4)
ms5=pd.Series(v5)
ms6=pd.Series(v6)
ms7=pd.Series(v7)
ms8=pd.Series(v8)
ms9=pd.Series(v9)
ms10=pd.Series(v10)
ms11=pd.Series(v11)
ms12=pd.Series(v12)

df_ventas_mes = pd.DataFrame( [ms1,ms2,ms3,ms4,ms5,ms6,ms7,ms8,ms9,ms10,ms11,ms12] )
df_ventas_mes
mayorm = df_ventas_mes.loc[df_ventas_mes["Ventas"].idxmax()]
menorm = df_ventas_mes.loc[df_ventas_mes["Ventas"].idxmin()]

promedio_mensual = df_ventas_mes["Ventas"].mean()

print(f"\nMes con mayor ventas: {mayorm['Mes']} ({mayorm['Ventas']})")
print(f"Mes con menor ventas: {menorm['Mes']} ({menorm['Ventas']})")
print(f"Venta promedio", promedio_mensual)
```

```python
#4.2
plt.figure(figsize=(5, 3))
plt.scatter(df_plnta ["Radio (km)"], df_plnta ["Masa (10^24 kg)"])
plt.title("Masa en Función del Radio")

for i, plnta in enumerate(df_plnta["Planeta"]):
    plt.text(df_plnta["Radio (km)"][i], df_plnta["Masa (10^24 kg)"][i], plnta, fontsize=7)

plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
#4.3
plt.figure(figsize=(5, 3))
plt.scatter(df_plnta['Periodo Orbital (días)'], df_plnta['Distancia al Sol (10^6 km)'])
plt.title("Distancia Sol vs Periodo Orbital")

for i, planeta in enumerate(df_plnta['Planeta']):
    plt.text(df_plnta["Periodo Orbital (días)"][i], df_plnta["Distancia al Sol (10^6 km)"][i], planeta, fontsize=5, ha='right')


plt.grid(axis='y', alpha=0.7)
plt.show()
```

```python
#4.4
df_planetas_s=df_plnta.sort_values(by="Gravedad (m/s^2)", ascending=False)

plt.figure(figsize=(5,3))
plt.bar(df_planetas_s["Planeta"], df_planetas_s["Gravedad (m/s^2)"])
plt.title("Gravedad Superficial de los Planetas")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.5)
plt.show()
```
