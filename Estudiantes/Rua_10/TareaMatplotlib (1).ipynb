````python
import matplotlib.pyplot as plt
import pandas as pd
````

````python
dic = {'Ana': 85, 'Juan': 92, 'María': 78, 'Pedro': 88, 'Lucía': 95}
dic
````

````python
dic['Carlos'] = 80
min_key = min(dic, key=dic.get)
del dic[min_key]
dic['María'] = 82
dic
````

````python
def encontrar_extremos(productos):
  producto_mas_caro = max(productos, key=productos.get)
  producto_mas_barato = min(productos, key=productos.get)
  return producto_mas_caro, producto_mas_barato
productos = {'manzanas': 2.5, 'naranjas': 3.0, 'plátanos': 1.2, 'peras': 2.8}
caro, barato = encontrar_extremos(productos)
print("Producto más caro:", caro)
print("Producto más barato:", barato)
````

````python
data = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'], 'Ventas': [2000, 2100, 1800, 2400, 2300, 2200, 2500, 2700, 2600, 2900, 3000, 3100]}
df_ventas = pd.DataFrame(data)
mes_mayor_venta = df_ventas['Mes'][df_ventas['Ventas'].idxmax()]
venta_mayor = df_ventas['Ventas'].max()
mes_menor_venta = df_ventas['Mes'][df_ventas['Ventas'].idxmin()]
venta_menor = df_ventas['Ventas'].min()
venta_promedio = df_ventas['Ventas'].mean()
print("Mes con mayor venta:", mes_mayor_venta, " - Ventas:", venta_mayor)
print("Mes con menor venta:", mes_menor_venta, " - Ventas:", venta_menor)
print("Venta promedio mensual:", venta_promedio)
````

````python
dic4 = {'Planeta': ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno'],
        'Masa (10^24 kg)': [0.330, 4.87, 5.97, 0.642, 1898, 568, 86.8, 102],
        'Radio (km)': [2439.7, 6051.8, 6371, 3389.5, 69911, 58232, 25362, 24622],
        'Distancia al Sol (10^6 km)': [57.9, 108.2, 149.6, 227.9, 778.5, 1434, 2871, 4495],
        'Periodo Orbital (días)': [88, 224.7, 365.2, 687, 4331, 10747, 30589, 59800],
        'Gravedad (m/s^2)': [3.7, 8.9, 9.8, 3.7, 24.8, 10.4, 8.7, 11.2]}
df_planetas = pd.DataFrame(dic4)
df_planetas
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df_planetas['Radio (km)'], df_planetas['Masa (10^24 kg)'])
plt.xlabel('Radio (km)')
plt.ylabel('Masa (10^24 kg)')
plt.title('Masa vs Radio de los Planetas')
plt.grid(True)
plt.show()
````

````python

plt.figure(figsize=(8, 6))
plt.scatter(df_planetas['Distancia al Sol (10^6 km)'], df_planetas['Periodo Orbital (días)'])
plt.xlabel('Distancia al Sol (10^6 km)')
plt.ylabel('Periodo Orbital (días)')
plt.title('Distancia al Sol vs Periodo Orbital de los Planetas')
plt.grid(True)
plt.show()
````

````python
df_planetas_ordenado = df_planetas.sort_values('Gravedad (m/s^2)', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(df_planetas_ordenado['Planeta'], df_planetas_ordenado['Gravedad (m/s^2)'])
plt.xlabel('Planeta')
plt.ylabel('Gravedad (m/s^2)')
plt.title('Gravedad Superficial de los Planetas (Ordenados)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````
