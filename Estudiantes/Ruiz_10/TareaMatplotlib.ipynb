````python
estudiantes = {'nombre': ['Ana', 'Juan', 'María', 'Pedro', 'Lucía'],
              'edad': [16, 17, 15, 16, 17],
              'grado': [10, 11, 10, 11, 12]}

df_estudiantes = pd.DataFrame(estudiantes)
print(df_estudiantes)
````

````python
datos_ventas = {'Mes': ['Enero', 'Febrero', 'Marzo', 'Abril', 'Mayo', 'Junio', 'Julio', 'Agosto', 'Septiembre', 'Octubre', 'Noviembre', 'Diciembre'],
                'Ventas': [2000, 2100, 1800, 2400, 2300, 2200, 2500, 2700, 2600, 2900, 3000, 3100]}

df_ventas = pd.DataFrame(datos_ventas)
mes_max_ventas = df_ventas.loc[df_ventas['Ventas'].idxmax()]
mes_min_ventas = df_ventas.loc[df_ventas['Ventas'].idxmin()]
venta_promedio = df_ventas['Ventas'].mean()
print("Mes con la mayor cantidad de ventas:")
print(mes_max_ventas)
print("\nMes con la menor cantidad de ventas:")
print(mes_min_ventas)
print("\nVenta promedio mensual:")
print(venta_promedio)
````

````python
planetas= {
    'Planeta': ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Júpiter', 'Saturno', 'Urano', 'Neptuno'],
    'masa': [0.330, 4.87, 5.97, 0.642, 1898, 568, 86.8, 102],
    'radio': [2439.7, 6051.8, 6371, 3389.5, 69911, 58232, 25362, 24622],
    'distancia_sol': [57.9, 108.2, 149.6, 227.9, 778.5, 1434, 2871, 4495],
    'periodo_orbital': [88, 224.7, 365.2, 687, 4331, 10747, 30589, 59800],
    'gravedad': [3.7, 8.9, 9.8, 3.7, 24.8, 10.4, 8.7, 11.2]}

df_planetas = pd.DataFrame(planetas)
print(df_planetas)
````

````python
import matplotlib.pyplot as plt
plt.figure(figsize=(8, 6))
plt.scatter(df_planetas['radio'], df_planetas['masa'])
plt.xlabel('Radio (km)')
plt.ylabel('Masa (10^24 kg)')
plt.title('Masa vs Radio de los Planetas')
plt.show()
````

````python
plt.figure(figsize=(8, 6))
plt.scatter(df_planetas['distancia_sol'], df_planetas['periodo_orbital'])
plt.xlabel('Distancia al Sol (millones de km)')
plt.ylabel('Periodo Orbital (días)')
plt.title('Distancia al Sol vs Periodo Orbital de los Planetas')
plt.show()
# existe una relacion entre las dos variables pues los planetas mas alejados del sol tienen periodos orbitales mas largos.
````

````python
df_planetas_ordenados = df_planetas.sort_values('gravedad', ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(df_planetas_ordenados['Planeta'], df_planetas_ordenados['gravedad'])
plt.xlabel('Planeta')
plt.ylabel('Gravedad Superficial (m/s^2)')
plt.title('Gravedad Superficial de los Planetas')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
````

````python
correlation_matrix = df_planetas[['masa', 'radio', 'distancia_sol', 'periodo_orbital', 'gravedad']].corr()
print(correlation_matrix)
import seaborn as sns
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', vmin=-1, vmax=1)
plt.title('Matriz de Correlación de Características de los Planetas')
plt.show()

#a masa y el radio tienen una correlación fuerte y positiva (0.99).
#La distancia al sol y el periodo orbital también tienen una correlación fuerte y positiva (0.99).
````
