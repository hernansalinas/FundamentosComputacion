### Entendiendo la presión atomosférica que siente un buzo.

Imagina que eres un buzo que entiende la física de los fluidos y quieres analizar cómo varía la presión con la profundidad bajo el agua. Entiendes que:
La presión hidrostática es la presión que se ejerce en un fluido en reposo debido a la fuerza de la gravedad. Esta presión aumenta con la profundidad debido al peso de la columna de agua que se encuentra por encima del punto considerado. La cual puede ser modelada con la siguiente expresión

$P=P_0+\rho gh $,

donde:

- $ P $: Presión total en el punto de profundidad \( h \) (en Pascales, Pa).
- $ P_0$: Presión atmosférica en la superficie (en Pascales, Pa).
- $ \rho $: Densidad del fluido (en kilogramos por metro cúbico, kg/m³).
- $ g $: Aceleración debida a la gravedad (aproximadamente 9.81 m/s² en la Tierra).
- $ h $: Profundidad bajo la superficie del fluido (en metros, m).

La anterior ecuación tiene varias implicaciones:


   - En la superficie, el buzo está sometido a la presión atmosférica estándar $ P_0 $
   - A medida que desciende, la presión adicional debido al agua se suma a la presión atmosférica, resultando en una presión total mayor.

   - Por ejemplo, a 10 metros de profundidad en agua dulce:
   $   P = 101325 + 1000  \times 9.81  \times 10 = 199125 \text{ Pa}$
  
   Esto equivale aproximadamente a 2 atmósferas de presión, considerando que 1 atm = 101325 Pa.

   - Cuando un buzo se sumerge en agua dulce en comparacion con  agua salada, existe una **diferencia** debido a la   **densidad**: El agua salada es más densa que el agua dulce debido a la salinidad, lo que implica que la presión hidrostática aumenta más rápidamente con la profundidad en el agua salada.
