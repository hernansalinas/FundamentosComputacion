# Comandos de interés de pandas

| Comando                      | Explicación                                                                                          |
|------------------------------|-----------------------------------------------------------------------------------------------------|
| `pd.DataFrame(data)`         | Crea un DataFrame a partir de un diccionario de datos o de una lista de listas, diccionarios, etc.  |
| `df['columna']`              | Selecciona una columna específica de un DataFrame.                                                  |
| `df[['col1', 'col2']]`       | Selecciona múltiples columnas específicas de un DataFrame.                                          |
| `df.loc[índice]`             | Selecciona una fila específica por índice, utilizando la etiqueta del índice.                      |
| `df.iloc[índice]`            | Selecciona una fila específica por índice, utilizando la posición numérica del índice.              |
| `df[df['columna'] > valor]`  | Filtra las filas del DataFrame donde los valores de una columna son mayores que un valor dado.      |
| `df['columna'].mean()`       | Calcula la media (promedio) de una columna numérica.                                               |
| `df['columna'].std()`        | Calcula la desviación estándar de una columna numérica.                                            |
| `df['columna'].min()`        | Encuentra el valor mínimo en una columna.                                                          |
| `df['columna'].max()`        | Encuentra el valor máximo en una columna.                                                          |
| `df.describe()`              | Genera estadísticas descriptivas para columnas numéricas, como la media, la desviación estándar, etc. |
| `df[df['columna'].notnull()]`| Filtra las filas del DataFrame donde los valores de una columna no son nulos.                      |
| `df.groupby('columna')`      | Agrupa el DataFrame por una o más columnas.                                                        |
| `df.groupby('columna').mean()` | Calcula la media de cada grupo para columnas numéricas después de agrupar por una columna.        |
| `df.plot(kind='bar')`        | Crea un gráfico de barras para los datos de un DataFrame.                                          |
| `df.plot(kind='scatter', x='col1', y='col2')` | Crea un gráfico de dispersión utilizando dos columnas especificadas.                   |
| `df.apply(func, axis=1)`     | Aplica una función a lo largo de un eje del DataFrame (por filas o columnas).                      |
| `df['columna'].astype(tipo)` | Cambia el tipo de datos de una columna específica.                                                 |
| `plt.xlabel('texto')`        | Establece la etiqueta del eje x en una gráfica de Matplotlib.                                      |
| `plt.ylabel('texto')`        | Establece la etiqueta del eje y en una gráfica de Matplotlib.                                      |
| `plt.title('texto')`         | Establece el título de una gráfica de Matplotlib.                                                  |
| `plt.grid(True)`             | Activa el grid (rejilla) en una gráfica de Matplotlib.                                             |
| `plt.show()`                 | Muestra la gráfica generada en Matplotlib.                                                         |

```python
def estado_materia(row):
  if row['Punto de Fusión (K)'] > 298 and row['Punto de Ebullición (K)'] > 298:
    return 'Sólido'
  elif row['Punto de Fusión (K)'] < 298 and row['Punto de Ebullición (K)'] > 298:
    return 'Líquido'
  else:
    return 'Gas'
df_elementos['Estado'] = df_elementos.apply(estado_materia, axis=1)
grupos_estado = df_elementos.groupby('Estado')
print(grupos_estado.groups)
```

```python
masa_atomica_promedio_por_estado = grupos_estado['Masa Atómica (u)'].mean()
print(masa_atomica_promedio_por_estado)
```

```python
masa_atomica_promedio_por_estado.plot(kind='bar')
plt.xlabel('Estado de la Materia')
plt.ylabel('Masa Atómica Promedio (u)')
plt.title('Comparación de Masas Atómicas Promedio por Estado')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
datos_particulas = {
    'Partícula': ['Electrón', 'Protón', 'Neutrón', 'Muón', 'Pión (+)', 'Pión (0)', 'Pión (-)', 'Kaón (K+)', 'Kaón (K0)'],
    'Masa (MeV/c^2)': [0.511, 938.272, 939.565, 105.658, 139.570, 134.977, 139.570, 493.677, 497.611],
    'Carga (e)': [-1, +1, 0, -1, +1, 0, -1, +1, 0],
    'Spin': ['1/2', '1/2', '1/2', '1/2', '0', '0', '0', '0', '0'],
    'Tiempo de Vida Media (s)': ['6.6 x 10^29', 'Estable', '881.5', '2.2 x 10^-6', '2.6 x 10^-8', '8.4 x 10^-17', '2.6 x 10^-8', '1.24 x 10^-8', '5.1 x 10^-8'],
    'Interacción Fundamental': ['Electromagnética', 'Electromagnética, Fuerte', 'Fuerte, Débil', 'Electromagnética', 'Fuerte', 'Fuerte', 'Fuerte', 'Fuerte', 'Fuerte']
}
df_particulas = pd.DataFrame(datos_particulas)
print(df_particulas)
```

```python
import matplotlib.pyplot as plt
tiempos_vida = []
for tiempo in df_particulas['Tiempo de Vida Media (s)']:
  if tiempo == 'Estable':
    tiempos_vida.append(float('inf'))
  else:
    tiempos_vida.append(float(tiempo.replace(' x 10^', 'e')))
plt.figure(figsize=(10, 6))
plt.scatter(df_particulas['Masa (MeV/c^2)'], tiempos_vida)
plt.xlabel('Masa (MeV/c^2)')
plt.ylabel('Tiempo de Vida Media (s)')
plt.title('Masa vs Tiempo de Vida Media de Partículas')
plt.yscale('log')
plt.grid(True)
plt.show()
```

```python
colores = {
    'Electromagnética': 'red',
    'Electromagnética, Fuerte': 'blue',
    'Fuerte, Débil': 'green',
    'Fuerte': 'purple'
}
plt.figure(figsize=(10, 6))
for interaccion, color in colores.items():
  df_subset = df_particulas[df_particulas['Interacción Fundamental'] == interaccion]
  plt.scatter(df_subset['Masa (MeV/c^2)'], df_subset['Carga (e)'], c=color, label=interaccion)
plt.xlabel('Masa (MeV/c^2)')
plt.ylabel('Carga (e)')
plt.title('Carga vs Masa de Partículas, Coloreadas por Interacción')
plt.legend()
plt.grid(True)
plt.show()
```

```python
conteo_spins = df_particulas['Spin'].value_counts()
plt.figure(figsize=(8, 6))
conteo_spins.plot(kind='bar')
plt.xlabel('Spin')
plt.ylabel('Número de Partículas')
plt.title('Distribución de Spin de Partículas')
plt.xticks(rotation=0)
plt.grid(True)
plt.show()
```
