````python
#4.2
plt.figure(figsize=(5, 3))
plt.scatter(df_planetas["Radio (km)"], df_planetas["Masa (10^24 kg)"])
plt.title("Masa en Función del Radio")

for i, planeta in enumerate(df_planetas["Planeta"]):
    plt.text(df_planetas["Radio (km)"][i], df_planetas["Masa (10^24 kg)"][i], planeta, fontsize=7)

plt.grid(axis='y', alpha=0.7)
plt.show()
````

````python
#4.3
plt.figure(figsize=(5, 3))
plt.scatter(df_planetas['Periodo Orbital (días)'], df_planetas['Distancia al Sol (10^6 km)'])
plt.title("Distancia Sol vs Periodo Orbital")

for i, planeta in enumerate(df_planetas['Planeta']):
    plt.text(df_planetas["Periodo Orbital (días)"][i], df_planetas["Distancia al Sol (10^6 km)"][i], planeta, fontsize=5, ha='right')


plt.grid(axis='y', alpha=0.7)
plt.show()
````

````python
#4.4
df_planetas_s=df_planetas.sort_values(by="Gravedad (m/s^2)", ascending=False)

plt.figure(figsize=(5,3))
plt.bar(df_planetas_s["Planeta"], df_planetas_s["Gravedad (m/s^2)"])
plt.title("Gravedad Superficial Planetas")
plt.xticks(rotation=45)
plt.grid(axis="y", alpha=0.5)
plt.show()
````
